#### 4. ☼ Read in the texts of the State of the Union addresses, using the `state_union` corpus reader. Count occurrences of `men`, `women`, and `people` in each document. What has happened to the usage of these words over time?

```python
nltk.corpus.state_union.fileids()
```

```python
for fileid in nltk.corpus.state_union.fileids():
    words = nltk.corpus.state_union.words(fileid)
    print(f"{fileid[:4]} {fileid[5:fileid.index(".")]} - men: {words.count("men")}, women: {words.count("women")}, people: {words.count("people")}")
```

```python
plt.figure(figsize=(12,8))
cfd = nltk.ConditionalFreqDist(
    (word, fileid[:4])
    for fileid in nltk.corpus.state_union.fileids()
    for word in nltk.corpus.state_union.words(fileid)
    if word in ["men", "women", "people"]
)
cfd.plot()
```

```python
print("Over time mention of women became as frequent as mention of men. There is very big variance in mention of people without any clear trends.")
```

#### 5. ☼ Investigate the holonym-meronym relations for some nouns. Remember that there are three kinds of holonym-meronym relation, so you need to use: `member_meronyms()`, `part_meronyms()`, `substance_meronyms()`, `member_holonyms()`, `part_holonyms()`, and `substance_holonyms()`.

```python
from nltk.corpus import wordnet as wn

for ss in wn.synsets("human"):
    print(ss.name())
    print(ss.definition())
```

```python
ss = wn.synset("homo.n.02")
print(f"Member holonyms of {ss.lemma_names()[0]}:")
for w in ss.member_holonyms():
    print("\t", w.lemma_names()[0])

print(f"Member meronyms of {ss.lemma_names()[0]}:")
for w in ss.member_meronyms():
    print("\t", w.lemma_names()[0])

print(f"Part holonyms of {ss.lemma_names()[0]}:")
for w in ss.part_holonyms():
    print("\t", w.lemma_names()[0])    

print(f"Part meronyms of {ss.lemma_names()[0]}:")
for w in ss.part_meronyms():
    print("\t", w.lemma_names()[0])

print(f"Substance holonyms of {ss.lemma_names()[0]}:")
for w in ss.substance_holonyms():
    print("\t", w.lemma_names()[0])

print(f"Substance meronyms of {ss.lemma_names()[0]}:")
for w in ss.substance_meronyms():
    print("\t", w.lemma_names()[0])
```

#### 6. ☼ In the discussion of comparative wordlists, we created an object called `translate` which you could look up using words in both German and Spanish in order to get corresponding words in English. What problem might arise with this approach? Can you suggest a way to avoid this problem?

```python
from nltk.corpus import swadesh

de2en = swadesh.entries(["de", "en"])
es2en = swadesh.entries(["es", "en"])
translate = dict(de2en)
translate.update(es2en)
print(f"Sum of entries is de2en and es2en {len(de2en) + len(es2en)}")
print(f"Number of entries in translate {len(translate)}")
```

```python
print("2 German words in de2en were replace with 2 Spanish words with the same spelling. This dictionaries should be kept separate to avoid ambiguity.")
```

#### 7. ☼ According to Strunk and White's Elements of Style, the word *however*, used at the start of a sentence, means "in whatever way" or "to whatever extent", and not "nevertheless". They give this example of correct usage: *However you advise him, he will probably do as he thinks best*. (http://www.bartleby.com/141/strunk3.html) Use the concordance tool to study actual usage of this word in the various texts we have been considering. See also the LanguageLog posting "Fossilized prejudices about 'however'" at http://itre.cis.upenn.edu/~myl/languagelog/archives/001913.html

```python
from nltk.corpus import brown

text = nltk.Text(brown.words())
text.concordance("However")
```

```python
from nltk.corpus import webtext

text = nltk.Text(webtext.words())
text.concordance("However")
```

```python
from nltk.corpus import gutenberg as gb

text = nltk.Text(gb.words())
text.concordance("However")
```

#### 8. ◑ Define a conditional frequency distribution over the Names corpus that allows you to see which *initial* letters are more frequent for males vs. females (cf. [4.4](https://www.nltk.org/book/ch02.html#fig-cfd-gender)).

```python
names = nltk.corpus.names
cdf = nltk.ConditionalFreqDist((fileid, word[0]) for fileid in names.fileids() for word in names.words(fileid))
plt.figure(figsize=(12, 8))
cdf.plot()
```
